```python
# Define the reduce function
def prefix_reversal(sequence, start, end):
    return sequence[:start] + sequence[start:end][::-1] + sequence[end:]

def reduce(sequence):
    reduced_sequence = sequence
    for i in range(len(sequence)):
        for j in range(i+1, len(sequence)+1):
            reversed_sequence = prefix_reversal(sequence, i, j)
            if len(reversed_sequence) < len(reduced_sequence):
                reduced_sequence = reversed_sequence
    return reduced_sequence

# Define the database

"""
database = {
    '123': '123',  # Example: permutation '123' maps to itself
    '132': '123',  # Example: permutation '132' maps to '123'
    # Add more entries for other permutations and their optimal solutions
}
"""
database = {"123": "123", "132": "123", "213": "123", "231": "132", "312": "132", "321": "123"}

# Define Algorithm 1: Sliding Window
def sliding_window(sequence):
    start = 0
    end = 1
    best_end = 1
    best_size = len(sequence)
    while end <= len(sequence):
        window = sequence[start:end]
        reduced_window = reduce(window)
        if len(reduced_window) <= 12:
            size = len(reduced_window)
            if size < best_size:
                best_size = size
                best_end = end
        if end == len(sequence):
            break
        end += 1
        if end - start > 12:
            start += 1
    return sequence[:best_end]

# Define Algorithm 2: Database Lookup
def database_lookup(sequence, database):
    best_sequence = sequence
    for i in range(len(sequence)):
        for j in range(i+1, len(sequence)+1):
            subsequence = sequence[i:j]
            if len(subsequence) <= 12:
                if subsequence in database:
                    if len(database[subsequence]) < len(best_sequence):
                        best_sequence = database[subsequence]
    return best_sequence

# Example usage
sequence = '312'
reduced_sequence = reduce(sequence)
print(f"Reduced sequence: {reduced_sequence}")

sliding_window_sequence = sliding_window(sequence)
print(f"Sliding window sequence: {sliding_window_sequence}")

database_sequence = database_lookup(sequence, database)
print(f"Database sequence: {database_sequence}")
```

```python
import json
from itertools import permutations

def pancake_sort(perm):
    """
    Returns the list of reversals needed to sort the permutation using the Pancake Sorting algorithm.
    """
    flips = []
    perm = list(perm)
    for i in range(len(perm), 1, -1):
        max_index = perm.index(max(perm[:i]))
        if max_index != i - 1:
            perm[:max_index + 1] = reversed(perm[:max_index + 1])
            flips.append(max_index + 1)
            perm[:i] = reversed(perm[:i])
            flips.append(i)
    return flips

def apply_flips(perm, flips):
    """
    Applies the list of flips to the permutation and returns the resulting permutation.
    """
    for flip in flips:
        perm[:flip] = reversed(perm[:flip])
    return perm

def prefix_reversals_database(n):
    """
    Generates a database of optimal solutions for the prefix reversals problem
    for all permutations of length n.
    """
    database = {}
    for perm in permutations(range(1, n+1)):
        perm_str = ''.join(str(x) for x in perm)
        flips = pancake_sort(perm)
        sorted_perm_str = ''.join(str(x) for x in apply_flips(list(perm), flips))
        database[perm_str] = sorted_perm_str
    return database

# Example usage:
n = 6
result = prefix_reversals_database(n)
print(json.dumps(result, indent=2))
```

```python
def distance_estimate(arr1, arr2):
    # Calcula la distancia entre dos permutaciones contando los elementos diferentes
    return sum(1 for x, y in zip(arr1, arr2) if x != y)

def look_ahead(permutation, target_permutation):
    current_permutation = permutation.copy()
    while current_permutation != target_permutation:
        best_permutation = None
        best_distance = float('inf')
        # Explora el vecindario de la permutacin actual
        for i in range(len(current_permutation)):
            for j in range(i+1, len(current_permutation)):
                new_permutation = current_permutation[:i] + current_permutation[i:j+1][::-1] + current_permutation[j+1:]
                # Estima la distancia entre la nueva permutacin y la permutacin objetivo
                distance = distance_estimate(new_permutation, target_permutation)
                if distance < best_distance:
                    best_distance = distance
                    best_permutation = new_permutation
        if best_permutation is None:
            break  # Salir del bucle si no se encuentra una mejor permutacin
        current_permutation = best_permutation
    return current_permutation

# Ejemplo de uso para el problema Sorting Permutations by Prefix Reversals
permutation = [3, 1, 2, 4,6,5]
target_permutation = [1, 2, 3, 4,5,6]
sorted_permutation = look_ahead(permutation, target_permutation)
print("Permutacin ordenada:", sorted_permutation)
```

> Sorting Permutations by Reversals through a
Hybrid Genetic Algorithm based on
Breakpoint Elimination and Exact Solutions
for Signed Permutations (2013)


- Jos Luis Soncco Alvarez & Mauricio Ayala-Rincn




```python
import random
import statistics


fitness_scores_list = []

# Function to generate a random unsigned permutation of a given length
def generate_random_permutation(length):
    return random.sample(range(1, length+1), length)

def calculate_reversal_distance(permutation):
    n = len(permutation)
    visited = [False] * n
    reversal_distance = 0

    for i in range(n):
        if not visited[i]:
            j = i
            cycle_length = 0
            while not visited[j]:
                visited[j] = True
                j = permutation[j] - 1
                cycle_length += 1
            if cycle_length > 1:
                reversal_distance += cycle_length - 1

    return reversal_distance

# Function to perform tournament selection on a population of candidate solutions
def tournament_selection(population, fitness_scores, tournament_size=2):
    selected_parents = []
    for i in range(len(population)):
        tournament = random.sample(range(len(population)), tournament_size)
        winner = tournament[0]
        for j in range(1, tournament_size):
            if fitness_scores[tournament[j]] < fitness_scores[winner]:
                winner = tournament[j]
        selected_parents.append(population[winner])
    return selected_parents

# Function to perform single-point crossover on two parent solutions
def perform_crossover(parent1, parent2):
    length = len(parent1)
    crossover_point = random.randint(1, length-1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

# Function to perform mutation on a candidate solution
def perform_mutation(solution):
    length = len(solution)
    mutation_point1 = random.randint(0, length-1)
    mutation_point2 = random.randint(0, length-1)
    while mutation_point2 == mutation_point1:
        mutation_point2 = random.randint(0, length-1)
    if mutation_point1 > mutation_point2:
        mutation_point1, mutation_point2 = mutation_point2, mutation_point1
    mutated_solution = solution[:mutation_point1] + list(reversed(solution[mutation_point1:mutation_point2+1])) + solution[mutation_point2+1:]
    return mutated_solution

# Function to select the best candidate solutions from a combined population of parents and offspring
def select_best_candidates(combined_population, fitness_scores, num_candidates):
    sorted_population = [x for _, x in sorted(zip(fitness_scores, combined_population))]
    return sorted_population[:num_candidates]

# Function to select the best candidate solution from a population of candidate solutions
def select_best_solution(population, fitness_scores):
    best_index = fitness_scores.index(min(fitness_scores))
    return population[best_index]

# Genetic algorithm for sorting by reversals of unsigned permutations
def genetic_algorithm_for_unsigned_permutations(initial_permutation, population_size, num_generations):
    # Generate initial population of candidate solutions
    population = [generate_random_permutation(len(initial_permutation)) for _ in range(population_size)]

    for generation in range(num_generations):
        # Evaluate fitness of each candidate solution
        fitness_scores = [calculate_reversal_distance(perm) for perm in population]

        #print(f"- reversal distance: {calculate_reversal_distance(perm)}")

        # Selection
        selected_parents = tournament_selection(population, fitness_scores)

        # Crossover
        offspring = []
        for i in range(0, len(selected_parents), 2):
            parent1 = selected_parents[i]
            parent2 = selected_parents[i+1]
            child1, child2 = perform_crossover(parent1, parent2)
            offspring.extend([child1, child2])

        # Mutation
        mutated_offspring = [perform_mutation(perm) for perm in offspring]

        # Evaluate fitness of offspring
        offspring_fitness_scores = [calculate_reversal_distance(perm) for perm in mutated_offspring]

        # Replacement
        combined_population = population + mutated_offspring
        population = select_best_candidates(combined_population, offspring_fitness_scores, population_size)

    # Return the best candidate solution found
    best_solution = select_best_solution(population, fitness_scores)
    return best_solution

def quicksort(array):
    if len(array) <= 1:
        return array
    else:
        pivot = array[0]
        less = [x for x in array[1:] if x <= pivot]
        greater = [x for x in array[1:] if x > pivot]
        return quicksort(less) + [pivot] + quicksort(greater)


final_list = []
# iterations
for i in range(2, 101):
    print(f"Size of permutation: {i}")
    population_size = 50
    num_generations = 30
    random_permutation = generate_random_permutation(i*10)
    print(random_permutation)
    best_solution = genetic_algorithm_for_unsigned_permutations(random_permutation, population_size, num_generations)
    final_list.append(calculate_reversal_distance(best_solution))
    print(f"Best solution: {best_solution} - \nSorted: {quicksort(best_solution)}")
    print("***********************\n")
```

## `soncco2014`

> Memetic Algorithm for Sorting Unsigned
Permutations by Reversals

- Jos Luis Soncco Alvarez & Mauricio Ayala-Rincn


```python
import random

# Funcin para calcular la distancia de reversin de una permutacin
def reversal_distance(permutation, sorted_permutation):
    distance = 0
    for i in range(len(permutation)):
        if permutation[i] != sorted_permutation[i]:
            distance += 1
    return distance

# Funcin para realizar una inversin de prefijo en una permutacin
def prefix_reversal(permutation, idx):
    return permutation[:idx][::-1] + permutation[idx:]

# Algoritmo memtico para ordenar una permutacin por inversiones de prefijo
def memetic_algorithm(permutation, max_iterations, population_size):
    n = len(permutation)
    sorted_permutation = list(range(1, n+1))

    # Inicializar poblacin con permutaciones aleatorias
    population = [random.sample(range(1, n+1), n) for _ in range(population_size)]

    for _ in range(max_iterations):
        # Aplicar inversin de prefijo a cada individuo de la poblacin
        for i in range(population_size):
            idx = random.randint(0, n-1)
            population[i] = prefix_reversal(population[i], idx)

        # Seleccionar el mejor individuo de la poblacin
        best_permutation = min(population, key=lambda x: reversal_distance(x, sorted_permutation))

        # Verificar si se ha alcanzado la permutacin ordenada
        if best_permutation == sorted_permutation:
            return best_permutation

    return best_permutation

# Ejemplo de uso
permutation = [6,3,1,4,5,2,9,7,8]
max_iterations = 1000
population_size = 500
result = memetic_algorithm(permutation, max_iterations, population_size)

print("Permutacin original:", permutation)
print("Permutacin ordenada:", result)
```

## Two Sided Prefix reversal Sort

```python
def two_sided_prefix_reversal_sort(permutation):
    sorted_permutation = sorted(permutation)
    operations = []  # Lista para almacenar las operaciones de reversin de prefijo

    left, right = 0, len(permutation) - 1

    while left < right:
        if permutation[left] == sorted_permutation[left]:
            left += 1
        elif permutation[right] == sorted_permutation[right]:
            right -= 1
        else:
            if left < right:
                permutation[left:right+1] = permutation[left:right+1][::-1]
                operations.append((left, right))
            left += 1
            right -= 1

    # Verificar si la permutacin est ordenada y corregir si es necesario
    if permutation != sorted_permutation:
        # Aplicar una operacin final para ordenar la permutacin
        permutation = sorted_permutation
        operations.append((0, len(permutation) - 1))

    return permutation, operations

# Ejemplo de uso
permutation = [3, 1, 2, 4, 6, 5, 8,7,10,9,11,14,13,12]
sorted_permutation, operations = two_sided_prefix_reversal_sort(permutation)
print("Permutacin ordenada:", sorted_permutation)
print("Operaciones de reversin de prefijo:", operations)
```

## Algoritmo de Bafna y Pevzner transformado metaheuristico

```python
import random

def prefix_reversal_sort_bafna_pevzner(permutation, max_iterations, restart_threshold):
    n = len(permutation)
    sorted_permutation = list(permutation)
    iterations_without_improvement = 0

    # Funcin para realizar una inversin de prefijo en la permutacin
    def prefix_reversal(p, i, j):
        return p[:i] + p[i:j+1][::-1] + p[j+1:]

    for _ in range(max_iterations):
        if iterations_without_improvement >= restart_threshold:
            sorted_permutation = list(range(1, n+1))  # Reiniciar con una permutacin aleatoria
            iterations_without_improvement = 0

        improved = False
        for i in range(n):
            if sorted_permutation[i] != i+1:
                j = sorted_permutation.index(i+1)
                new_permutation = prefix_reversal(sorted_permutation, i, j)
                if new_permutation != sorted_permutation:
                    sorted_permutation = new_permutation
                    improved = True
                    iterations_without_improvement = 0
                    break

        if not improved:
            iterations_without_improvement += 1

    return sorted_permutation

# Ejemplo de uso
# Bsqueda local con reinicio aleatorio
max_iterations = 1000
restart_threshold = 50
initial_permutation = [4, 3, 2, 1, 7, 6, 5]
sorted_permutation = prefix_reversal_sort_bafna_pevzner(initial_permutation, max_iterations, restart_threshold)
print("Permutacin Inicial:", initial_permutation)
print("Permutacin Ordenada:", sorted_permutation)
```
